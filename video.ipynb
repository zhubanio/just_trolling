```python
import cv2
frame = cv2.imread("just_trolling/ltg.jpg", 1)
frame.shape
```

```python
from matplotlib import pyplot as plt
plt.imshow(frame)
plt.show()
cv2.imwrite("just_trolling/blue_ltg.jpg", frame)
```

```python
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) #  
cv2.imwrite("just_trolling/blue_ltg.jpg", frame)
```

```python
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) #  
plt.imshow(frame)
plt.show()
```

```python
scale_percent = 10
print(frame.shape)
height = int(frame.shape[0] * scale_percent / 100)
width = int(frame.shape[1] * scale_percent / 100)
dim = (width, height)
resized = cv2.resize(frame,  dim, interpolation = cv2.INTER_AREA)
print(resized.shape)
plt.imshow(resized)
plt.show()
```

```python
(h, w, c) = frame.shape
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated = cv2.warpAffine(frame, M, (w, h))

plt.imshow(rotated)
plt.show()
```
